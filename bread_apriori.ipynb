```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('bread basket.csv')
print("DataFrame shape: ", data.shape)
data.head()
```

```python
#format waktu
data['date_time'] = pd.to_datetime(data['date_time'], format="%d-%m-%Y %H:%M")
```

```python
data["date_time"].dtype 
```

```python
data["month"] = data["date_time"].dt.month
data["day"] = data["date_time"].dt.weekday
data["hour"] = data["date_time"].dt.hour
data.head()
```

```python
 #menampilkan data 10 item terlaris

plt.figure(figsize=(13,5))
sns.set_palette("muted")

sns.barplot(x=data['Item'].value_counts()[:10].index,
            y=data['Item'].value_counts()[:10].values )
plt.xlabel("")
plt.ylabel("")
plt.xticks(size=13,rotation=45)
plt.title("Top 10 Items Sold",size=17)
plt.show()
```

```python
#transaksi tiap bulan

data_perbulan = data.groupby("month")["Transaction"].count()
data_perbulan = pd.concat([data_perbulan.iloc[4:],data_perbulan.iloc[:4]])

plt.figure(figsize=(8,5))
sns.barplot(x=["October","November","December","January","February","March","April"],
            y=data_perbulan.values, color="skyblue")
plt.xticks(size=13)
plt.title("Total Transactions per Month",size=17)

plt.show()
```

```python
#transaksi perhari

data_perhari = data.groupby("day")["Transaction"].count()
data_perhari = pd.concat([data_perhari.iloc[4:],data_perhari.iloc[:4]])

plt.figure(figsize=(8,5))
sns.barplot(x=["Senin","Selasa","Rabu","Kamis","Jumat","Sabtu","Minggu"],
            y=data_perhari.values, color="#ff9999")
plt.xticks(size=13)
plt.title("Total Transactions per Hari",size=17)

plt.show()
```

```python
#transaksi perhjam

data_perjam = data.groupby("hour")["Transaction"].count()
data_perjam = pd.concat([data_perjam.iloc[4:],data_perjam.iloc[:4]])

plt.figure(figsize=(8,5))
sns.barplot(x=data_perjam.index,
            y=data_perjam.values, color="#ffcc99")
plt.xticks(size=13)
plt.title("Total Transactions per Hari",size=17)

plt.show()
```

```python
#data prep
data["Item"] = data["Item"].apply(lambda item: item.lower())
```

```python
data["Item"] = data["Item"].apply(lambda item: item.strip())
```

```python
data = data[["Transaction","Item"]].copy()
data.head(10)
```

```python
from mlxtend.frequent_patterns import association_rules, apriori
```

```python
item_count = data.groupby(["Transaction","Item"]).size().reset_index(name="Count")
item_count.head(10)
```

```python
item_count_pivot = item_count.pivot_table(index='Transaction',columns='Item',values='Count',aggfunc="sum").fillna(0)
print("Pivot Table shape: ", item_count_pivot.shape)
item_count_pivot.head()
```

```python
item_count_pivot = item_count_pivot.astype("int32")
item_count_pivot.head()
```

```python
def encode_units(x):
    if x <= 0:
        return 0
    if x >= 1:
        return 1

item_count_pivot = item_count_pivot.applymap(encode_units)
item_count_pivot.head()
```

```python
print("Ukuran Dataset: ", item_count_pivot.shape)
print("Jumlah Transaksi: ", item_count_pivot.shape[0])
print("Jumlah Item: ", item_count_pivot.shape[1])
```

```python
support = 0.01
frequent_itemsets = apriori(item_count_pivot, min_support=support, use_colnames=True)
frequent_itemsets.sort_values("support",ascending=False).head(10)
```

```python
metric = "lift"
min_threshold = 1

rules = association_rules(frequent_itemsets, metric=metric, min_threshold=min_threshold)[['antecedents', 'consequents', 'support', 'confidence', 'lift']]
rules.sort_values('confidence',ascending=False,inplace=True)
rules.head(15)
```
